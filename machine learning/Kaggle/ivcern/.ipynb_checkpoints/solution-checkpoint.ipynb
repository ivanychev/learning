## Metrics

The competition holders offer a tricky metric called AMS:

$$
\mathrm{AMS} = \sqrt{2\left((s + b + b_{\text{reg}})\log\left(1 + \frac{s}{b + b_{\text{reg}}}\right) - s\right)}
$$

$b_{\text{reg}}$ is known regularizing constant. $s$ and $b$ are True Positive Rate and False Positive Rate respectively, which could be computed as

$$
s = \sum_{i=1}^l w_i [y_{\text{true}} = 1]\cdot[y_{\text{pred}} = 1]
$$
$$
b = \sum_{i=1}^l w_i [y_{\text{true}} = 0]\cdot[y_{\text{pred}} = 1]
$$

$w_i$ are the weights provided in training dataset.

```python
B_R = 10.0
NORMALIZER = 550000.0

def raw_TPR(y_true, y_pred, weights):
    return np.sum(weights * ((y_true == 1) & (y_pred == y_true)))

def raw_FPR(y_true, y_pred, weights):
    return np.sum(weights * ((y_true == 0) & (y_pred != y_true)))

def AMS(y_true, y_pred, weights):
    tpr = raw_TPR(y_true, y_pred, weights)
    fpr = raw_FPR(y_true, y_pred, weights)
    score = np.sqrt(NORMALIZER / y_pred.size) * np.sqrt(2 * ((tpr + fpr + B_R) * np.log(1 + tpr / (fpr + B_R))) - tpr)
    return score
```

## Choosing of model. XGBoost.

I propose to deal with our task with gradient boosting approach. This method tend to efficiently balance between prediction accuracy and model complexity. Big advantage is robustness against NAs in dataset, so we don't need any processing of empty cells. Here we will use [XGBoost library](https://github.com/dmlc/xgboost) because its highly optimized, multithread and it tends to outperform other GradientBoosting implementations (i.e. `sklearn.ensemble.GradientBoostingClassifier`).


Here are some scripts that will help to save the solution and generate rankings according to the predicted probability.

```python
def make_rankings(probs, treshold):
    num = probs.shape[0]
    arg_sorted = np.argsort(probs)
    rng = np.arange(1, num + 1)
    probs_order = np.zeros(num)
    probs_order[arg_sorted] = rng
    return probs_order

def save_solution_from_prob(ids, y_prob, filename, treshold):
    with open(filename, "w") as f:
        y_labels = list(map(lambda x: "s" if x >= treshold else "b", list(y_prob)))
        y_rank = make_rankings(y_prob, treshold)
        f.write("EventId,RankOrder,Class\n")
        for i in range(len(y_labels)):
            f.write("%d,%d,%s\n" % (ids[i], y_rank[i], y_labels[i]))

def generate_solution(cls, X_test, indexes, name="sol.csv", treshold=0.5):
    y_prob = cls.predict_proba(X_test)[:, 1]
    save_solution_from_prob(indexes, y_prob, name, treshold)
```

## Grid searching XGBoost

To find the best solution we will search through the grid of parameters. `ParameterGrid` from `sklearn.grid_search` is used to implement this. Due to the computational complexity of such experiment I have learned how to use remote servers to perform this kind of work. I used Amazon EC2 vm (scripts that enabled me to quickly install required components for Ubuntu 14.04 are available at the root directory of current project). 

_Following code is commented out because execution will take an enormous amount of time. This code is used to find optimal parameters for XGBoost._

```python
METRIC = 'ams@0.14'
THRESHOLD = 0.14

def train_xgb(param, X, y, weights, missing=NA_VALUE):
    xgmat = xgb.DMatrix(X, label=y, missing=missing, weight=weights)
    param['objective'] = 'binary:logitraw'
    param['scale_pos_weight'] = weights[y == 0].sum() / weights[y == 1].sum()
    param['eval_metric'] = 'ams@0.14'
    num_rounds  = param['num_rounds']
    param['silent'] = 1
    plst = list(param.items())
    bst = xgb.train(plst, xgmat, num_rounds)
    return bst

def get_pred_from_proba(proba, threshold):
    cutoff_level = sorted(proba, reverse=True)[int(proba.size * threshold)]
    y_pred = np.zeros(proba.size)
    for i in range(y_pred.size):
        y_pred[i] = 1 if proba[i] >= cutoff_level else 0
    return y_pred

def get_cv_scores(param, cv, X, y, weights, missing=NA_VALUE, threshold=THRESHOLD):
    scores = np.zeros(cv.n_folds)
    idx = 0
    for train, test in cv:
        bst = train_xgb(param, X[train], y[train], weights[train])
        xgmat_pred = xgb.DMatrix(X[test], missing=missing)
        y_pred_proba = bst.predict(xgmat_pred)
        y_pred = get_pred_from_proba(y_pred_proba, threshold)
        score = AMS(y[test], y_pred, weights[test])
        scores[idx] = score
        idx += 1
    return scores
```

```python
param_grid = {
    'bst:max_depth':[6, 10, 13],
    'bst:eta':[0.01,0.1, 0.5],
    'bst:subsample':[0, 0.5, 1],
    "num_rounds":[300,1000, 3000]
}

grid = ParameterGrid(param_grid)
cv = KFold(y_train.size, 5)
max_score = None
max_param = None
for param in grid:
    scores = get_cv_scores(param, cv, X_train, y_train, weights)
    score = scores.mean()
    print(param, score)
    if max_score is None or score > max_score:
        max_param = param
```

## Printing out results

```python
threshold_ratio = 0.15
outfile = "xgb_res.csv"

res  = [ (int(index_test[i]), y_pred[i] ) for i in range(len(y_pred)) ] 

rorder = {}
rorder_len = 0
for k, v in sorted(res, key = lambda x: x[1], reverse=True):
    rorder[k] = rorder_len + 1
    rorder_len += 1


ntop = int(threshold_ratio * rorder_len)
fo = open(outfile, 'w')
nhit = 0
ntot = 0
fo.write('EventId,RankOrder,Class\n')
for k, v in res:        
    if rorder[k] <= ntop:
        lb = 's'
        nhit += 1
    else:
        lb = 'b'        

    fo.write('%s,%d,%s\n' % ( k,  len(rorder)+1-rorder[k], lb ) )
    ntot += 1
fo.close()
```
