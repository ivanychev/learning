> Загрузите картинку parrots.jpg. Преобразуйте изображение, приведя все значения в интервал от 0 до 1. Для этого можно воспользоваться функцией img_as_float из модуля skimage.

> Создайте матрицу объекты-признаки: характеризуйте каждый пиксель тремя координатами - значениями интенсивности в пространстве RGB.

````python
image = img_as_float(image)
````

> Запустите алгоритм K-Means с параметрами init='k-means++' и random_state=241. После выделения кластеров все пиксели, отнесенные в один кластер, попробуйте заполнить двумя способами: медианным и средним цветом по кластеру.

````python
def generate_class_pixel_arrays(clr, X):
    labels_counter = Counter(clr.labels_)
    labels_pixels = []
    for key in labels_counter.keys():
        arr = X[clr.labels_ == key]
        labels_pixels.append(arr)
    return labels_pixels, labels_counter

def get_means_and_medians(labels_pixels):
    labels_medians = []
    labels_means = []
    for pixel_arr in labels_pixels:
        labels_means.append(pixel_arr.mean(axis=0))
        labels_medians.append(np.median(pixel_arr, axis=0))
    return (labels_medians, labels_means)

def paste_means_and_medians(clr, X, labels_num, labels_means, labels_medians):
    X_medians = X.copy()
    X_means = X.copy()

    for key in labels_num.keys():
        X_means[clr.labels_ == key] = labels_means[key]
        X_medians[clr.labels_ == key] = labels_medians[key]

    image_medians = X_medians.reshape(image.shape)
    image_means = X_means.reshape(image.shape)
    return (image_medians, image_means)

def picture_do(clusters, image):
    X = image.reshape(474 * 713, 3)
    clr = KMeans(init='k-means++', random_state=241, n_clusters=clusters)
    clr.fit(X)
    
    labels_pixels, labels_counter = generate_class_pixel_arrays(clr, X)
    labels_medians, labels_means = get_means_and_medians(labels_pixels)
    image_medians, image_means = paste_means_and_medians(clr, X, labels_counter, labels_means, labels_medians)
    return (image_medians, image_means)
    
````

````python
def MSE(I, K):
    assert I.shape == K.shape
    # ret = 1 / (I.shape[0] * I.shape[1]) * np.sum(np.power((I - K), 2))
    ret = np.power((I - K), 2).mean()
    return ret

def PSNR(I, K):
    assert I.shape == K.shape
    return 10 * np.log10(1 / MSE(I, K))
````
